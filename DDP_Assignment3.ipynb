### What is Structured Streaming
Structured Streaming is a scalable stream processing engine built on top of the Spark SQL engine. It enables developers to process continuous streams of data in real time using the same APIs that are used for batch processing in Spark.

In Structured Streaming, streaming data is treated as an unbounded table that continuously grows as new data arrives. Spark processes this data incrementally and updates the results in real time.

### Difference Between Batch Processing and Streaming
Batch processing involves processing a fixed dataset that already exists in storage. The computation starts after the complete dataset is available.

Streaming processing works on continuously generated data such as logs, sensor data, financial transactions, and social media feeds. Instead of waiting for all data to arrive, the system processes incoming data immediately.

### Key Features of Structured Streaming
1. High scalability through distributed processing
2. Fault tolerance using checkpointing and write‑ahead logs
3. Unified API for both batch and streaming processing
4. Integration with Spark SQL and DataFrame operations
5. Support for multiple streaming sources such as Kafka, file systems, and sockets

### Real‑Time Use Cases
Structured Streaming is widely used in real‑time analytics systems. Some common applications include fraud detection in financial transactions, monitoring server logs for anomalies, analyzing IoT sensor data, and building real‑time recommendation systems.

## Part (b)
### Step 1: Create Spark Session

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, split

spark = SparkSession.builder.appName("StructuredStreamingWordCount").getOrCreate()

spark
```

### Step 2: Read Streaming Data From Socket

```python
lines = spark.readStream.format("socket").option("host","localhost").option("port",9999).load()

lines
```

### Step 3: Split Words From Incoming Stream

```python
words = lines.select(explode(split(lines.value," ")).alias("word"))

words
```

### Step 4: Perform Word Count

```python
wordCounts = words.groupBy("word").count()

wordCounts
```

### Step 5: Output Streaming Results to Console

```python
query = wordCounts.writeStream \
    .outputMode("complete") \
    .format("memory") \
    .queryName("word_counts_in_memory") \
    .start()
```
