### Distributed Data Processing
Distributed data processing is a computing approach used to process extremely large datasets by dividing the workload across multiple machines that work together as a cluster. Instead of relying on a single computer to perform all computations, the dataset is partitioned into smaller chunks and distributed among several worker nodes. Each node processes its portion of the data simultaneously, and the results are later combined to produce the final output.

This approach is necessary when datasets become too large to be processed efficiently on a single machine. Distributed systems allow organizations to scale horizontally by adding more machines to the cluster instead of increasing the power of a single system.

The main advantages of distributed data processing include improved scalability, faster processing through parallelism, and better fault tolerance. If one machine fails, other machines in the cluster can continue the computation.

### Apache Spark and Performance Improvement
Apache Spark is a distributed data processing framework designed for high‑performance analytics on large-scale datasets. It improves performance compared to traditional data processing frameworks such as Hadoop MapReduce.

In Hadoop MapReduce, each step of computation writes intermediate results to disk before moving to the next step. This frequent disk I/O significantly slows down processing.

Spark improves performance by executing many operations in memory rather than writing intermediate results to disk. This dramatically reduces latency and increases speed, especially for iterative algorithms and interactive analytics.

Spark also uses a Directed Acyclic Graph (DAG) execution engine that optimizes the entire sequence of operations before executing them. This optimization minimizes data shuffling and improves efficiency.

### Role of In‑Memory Computation
In‑memory computation is one of the most important features of Apache Spark. Instead of repeatedly reading data from disk, Spark stores intermediate datasets in memory whenever possible.

This is particularly useful for machine learning algorithms, graph processing, and iterative computations where the same dataset is accessed multiple times. By avoiding disk access, Spark can perform these operations significantly faster.

Because of this capability, Spark can often run workloads up to 10–100 times faster than traditional disk‑based processing frameworks.

### Step 1: Create Spark Session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkAssignment1").getOrCreate()

spark
```

### Step 2: Create Sample Employee Data

```python
data = [
    (1,"Alice",25,35000),
    (2,"Bob",30,28000),
    (3,"Charlie",29,42000),
    (4,"David",35,50000),
    (5,"Eva",24,26000)
]

columns = ["ID","Name","Age","Salary"]

df = spark.createDataFrame(data, columns)

df
```
