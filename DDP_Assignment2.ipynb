### Transformations in Apache Spark
Transformations are operations applied to datasets that create a new dataset from an existing one. Transformations are lazy operations, meaning that they are not executed immediately when they are defined. Instead, Spark records the transformations and builds a logical execution plan called a lineage graph.

Spark only executes these transformations when an action is called. This strategy allows Spark to optimize the execution plan before running the computation.

Examples of transformations include:

1. **select()** – used to retrieve specific columns from a dataset.
2. **filter()** – used to retrieve rows that satisfy a given condition.
3. **groupBy()** – used to group data based on a column so that aggregation operations such as count or sum can be performed.

### Actions in Apache Spark
Actions are operations that trigger the execution of transformations and return results. When an action is called, Spark executes the entire lineage of transformations and computes the result.

Examples of actions include:

1. **show()** – displays the contents of a DataFrame.
2. **count()** – returns the number of rows in a dataset.
3. **collect()** – retrieves all records from the distributed cluster to the driver program.

### Importance of Transformations and Actions
The separation of transformations and actions enables Spark to perform optimization before execution. Since transformations are lazy, Spark can analyze the entire computation pipeline and optimize it using techniques such as pipelining and minimizing data shuffling.

This improves performance and resource utilization in distributed environments.

## Part (b)
### Step 1: Create Spark Session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkAssignment2").getOrCreate()

spark
```

### Step 2: Create Dataset

```python
data = [
    (1,"Alice","IT",50000),
    (2,"Bob","HR",42000),
    (3,"Charlie","IT",55000),
    (4,"David","Finance",48000),
    (5,"Eva","HR",40000),
    (6,"Frank","IT",60000)
]

columns = ["ID","Name","Department","Salary"]

df = spark.createDataFrame(data, columns)

df.show()
```

### Step 3: Select Specific Columns

```python
selected_df = df.select("Name","Department")

selected_df.show()
```

### Step 4: Apply Filter Condition

```python
filtered_df = df.filter(df.Salary > 45000)

filtered_df.show()
```

### Step 5: Group Data and Calculate Count

```python
grouped_df = df.groupBy("Department").count()

grouped_df.show()
```
